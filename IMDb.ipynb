# Outlier Detection Using the IQR Method

Outliers were identified in the numeric columns using the Interquartile Range (IQR) method:

```python
# Select numeric columns for outlier analysis
numeric_cols = dataframe.select_dtypes(include=['float64', 'int64']).columns
print("Numeric Columns:", numeric_cols)

for col in numeric_cols:

# Detect outliers using IQR method
    Q1 = dataframe[col].quantile(0.25)
    Q3 = dataframe[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = dataframe[(dataframe[col] < lower) | (dataframe[col] > upper)]
    print(f"{col} -> {len(outliers)} outliers")
```

This code calculates the first (Q1) and third (Q3) quartiles to find the IQR.
Any data points lying below Q1 - 1.5×IQR or above Q3 + 1.5×IQR are considered outliers.

#### Purpose:

Detecting outliers helps identify unusual or extreme values in numerical features such as ratings or duration.

It’s useful for understanding data distribution and improving data quality before visualization or modeling.
### Outlier Detection Results
After applying the IQR method:
- Year → 0 outliers
- Rating → 8 outliers

#### Overview:

The Year column shows no outliers, meaning all movie release years fall within a consistent range.
However, the Rating column has 8 outliers, which correspond to movies with exceptionally high IMDb ratings.

#### Insight:

These outliers are not errors but represent top-rated films (around 9.0–9.3).
In this context, such outliers are meaningful and highlight exceptional movie quality rather than data issues.

# Converting Movie Duration into Total Minutes

The following code converts the movie duration from text format (like "2h 39m") into total minutes:

```python
# Convert duration like "2h 39m" into total minutes

def convert_duration(duration):
    # Start with 0 hours and 0 minutes
    hours = 0
    minutes = 0

    # If there's an 'h', split and take the number before it
    if 'h' in duration:
        hours = int(duration.split('h')[0].strip())

    # If there's an 'm', take the number before 'm'
    if 'm' in duration:
        # If both h and m exist, split after 'h'
        if 'h' in duration:
            minutes = int(duration.split('h')[1].replace('m', '').strip())
        else:
            minutes = int(duration.replace('m', '').strip())

    # Total minutes
    total_minutes = hours * 60 + minutes
    return total_minutes

# Apply the function
df['Duration'] = df['Duration'].apply(convert_duration)

```

#### Explanation:
- The function extracts hours and minutes separately from strings like "2h 39m".
- It then converts them into integers and calculates the total time in minutes.
- The transformation is applied to the entire Duration column using .apply().
#### Purpose:
This makes the Duration column numerical, allowing easier statistical analysis and visualizations such as average runtime or comparisons of movie lengths.

# Creating the Rating_Category Column

A new column named “Rating_Category” is created using:

```python
def rating_category(r):
    if r >= 9:
        return 'Outstanding'
    elif r >= 8.5:
        return 'Excellent'
    elif r >= 8.0:
        return 'Very Good'
    else:
        return 'Good'

df['Rating_Category'] = df['Rating'].apply(rating_category)
```

#### Explanation:
- This function classifies each movie based on its IMDb rating.
- Movies with a rating of 9 or above are tagged as Outstanding.
- Ratings between 8.5 and 8.9 are labeled Excellent.
- Ratings between 8.0 and 8.4 are Very Good.
- Anything below 8.0 falls under Good.
#### Purpose:
This categorization makes it easier to analyze movie quality groups and compare trends in high-rated films.

# Creating the Is_Long Column

A new column named “Is_Long” is added using:

```python
df['Is_Long'] = df['Duration'].apply(lambda x: "Yes" if x > 150 else "No")
```

#### Explanation:
- Each movie’s duration is checked.
- If the duration is greater than 150 minutes, it’s labeled as “Yes”, otherwise “No”.
#### Purpose:
This classification helps compare long movies (over 2.5 hours) with shorter ones to see if runtime affects ratings or popularity.

# Creating the Decade Column

A new column named “Decade” is created using:

```python
df['Decade'] = (df['Year'] // 10) * 10
```

- Each movie’s release year is divided by 10, then multiplied by 10 to group it into its decade (for example, 1994 → 1990, 2003 → 2000).
- This helps categorize movies based on the decade they were released.

#### Purpose:
Grouping by decade makes it easier to analyze trends in movie production and ratings over time — for example, identifying which decade produced the most highly rated films.

# Simplifying Certificate Categories for Consistency

The following code standardizes the Certificate column to make movie ratings more uniform and easier to analyze:

```python
# Simplify Certificate categories for consistency
df['Certificate_Cleaned'] = df['Certificate'].replace({
    'Approved': 'General',
    'Passed': 'General',
    'G': 'General',
    'PG': 'Parental Guidance',
    'PG-13': 'Teens',
    'R': 'Adults',
    'NC-17': 'Adults Only',
    'Unrated': 'Unrated',
    'Not Rated': 'Unrated'
})
```

#### Explanation:
- Different certificate labels that mean the same thing are grouped together.
- For example, Approved, Passed, and G are all renamed as General.
- PG becomes Parental Guidance, and PG-13 becomes Teens.
- Adult-oriented ratings like R and NC-17 are unified as Adults and Adults Only.
- Missing or unclear values such as Unrated and Not Rated are kept consistent as Unrated.
#### Purpose:
This step ensures cleaner and more consistent categories, making it easier to perform group analysis or visualizations by movie rating types.
